```python
class BankAccount:
  def __init__(self, account_number):
    self.account_number = account_number
  def deposit(self, amount):
    print(f"Depositing ${amount} to Account Number : {self.account_number}")
  def withdraw(self, amount):
    print(f'Withdrawing $ {amount} from Acount Number : {self.account_number}')

# Create an object of the BankAccount class
account = BankAccount("1234567890")

# Accessing object attributes and methods
print(account.account_number)

account.deposit(1000) # Output : Depositing $1000 to Account Number : 1234567890
account.withdraw(50) # Output : Withdrawing $500 from Account Number : 1234567890
```

```python
import unittest

# User-defined class
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def compute_area(self):
        return 3.14 * self.radius**2


# Class to test Circle
class TestCircle(unittest.TestCase):
    def test_compute_area(self):
        c = Circle(0)
        self.assertEqual(c.compute_area(), 0.0)

        c = Circle(5)
        self.assertEqual(c.compute_area(), 78.5)

    def test_will_fail(self):
        c = Circle(5)
        self.assertLess(c.compute_area(), 0)

if __name__ == "__main__":
    unittest.main()
```

```python

import unittest

def evens(numbers):
    """Return the even values in numbers"""
    return [i for i in numbers if (i % 2 == 0)]

def odds(numbers):
    """Return the odd values in numbers"""
    return [i for i in numbers if (i % 2 == 1)]

test_nums = [1, 3, 5, 6, 8, 2, 1]
class TestNumbers(unittest.TestCase):
    # We can also have the test_nums as a local variable in the class but we would need to call it using self.test_nums
    def test_evens(self):
        even_list = evens(test_nums)
        for i in even_list:
            self.assertEqual(i%2,0)

    def test_odds(self):
        odd_list = odds(test_nums)
        for i in odd_list:
            self.assertNotEqual(i%2,0)


if __name__ == "__main__":
    unittest.main()

# This is for the test case to find out if the output is really purely even or odd numbers.
```
